```python
from MetricsExp import TKPrometheus

import pandas as pd
import os

col = ["name","start","end","requests","type"]
col_metrics = ["name", "url","level"]
node_interface_instance_col = ["node_name","interface","ip","node_ip"]
res_col = ["function", "invoke_time", "time_model_load", "startTime", "endTime", "runtime","ip","response_time","request_vol","interface","type","token"]

res_path = '/home/tinker/workspace/Euler-Serverless/Platform-Testing/version0818'
# function_res  = read all csv file end with -result.csv
function_res = pd.DataFrame()
for file in os.listdir(res_path):
    if file.endswith('-result.csv'):
        function_res = pd.concat([function_res,pd.read_csv(res_path+'/'+file, names=res_col, header=None)])

function_run = pd.read_csv('/home/tinker/workspace/Euler-Serverless/Platform-Testing/version0818/record.csv',names=col,header=None)

metric = pd.read_csv('/home/tinker/workspace/Euler-Serverless/Platform-Testing/metric/metric.csv',names=col_metrics,header=None,sep=',%')

interface = pd.read_csv("/home/tinker/workspace/Euler-Serverless/Platform-Testing/node_interface_instance.csv", names=node_interface_instance_col)

metric
# function_run
interface
```

```python
function_res
function_res['ip'] = function_res['ip'].apply(lambda x: x.split('.')[0] + '.' + x.split('.')[1] + '.' + x.split('.')[2] + '.' + '0')
# function_res.merge(interface, de, how='left')
function_res = pd.merge(function_res, interface, on='ip', how='left')
function_res
```

```python
prom = TKPrometheus()
columns_all=['requests','type','timestamp','value','metrics','node_ip','function','interface','level']
df = pd.DataFrame(columns=columns_all)
for index, f in function_res.iterrows():
    for i,m in metric.iterrows():
        fet = prom.fetch_grafana(start=f['invoke_time'],end = f['response_time'], query=m['url'],step=0.5, name=m['name'], function=f['function'],req =f['request_vol'],type=f['type'],node_ip=f['node_ip'],interface=f['interface_x'])
        print(f)
        fet['level'] = m['level']
        df = pd.concat([df,fet])
df.to_csv('/home/tinker/workspace/Euler-Serverless/Platform-Testing/version0818/'+ 'Kmetrics.csv', mode='a', header=False, index=False)
```
